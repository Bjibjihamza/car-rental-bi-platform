```python
import os
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("ggplot")

DATA_DIR = "data"  # adapte selon ton arborescence
```

```python
def load_all_telemetry(data_dir=DATA_DIR):
    files = sorted(glob.glob(os.path.join(data_dir, "iot_telemetry_*.csv")))
    if not files:
        raise FileNotFoundError("Aucun fichier CSV trouvÃ© dans data/. GÃ©nÃ¨re la tÃ©lÃ©mÃ©trie d'abord.")
    
    dfs = []
    for f in files:
        df = pd.read_csv(f, parse_dates=["TIMESTAMP"])
        df["SOURCE_FILE"] = os.path.basename(f)
        dfs.append(df)

    full_df = pd.concat(dfs, ignore_index=True)
    print(f"ðŸ“¦ {len(files)} fichiers chargÃ©s, {len(full_df):,} lignes au total")
    return full_df

df = load_all_telemetry()
df.head()
```

```python
import numpy as np
from IPython.display import display

def check_start_stop(df, tolerance=0.0005):
    """
    tolerance = tolÃ©rance en latitude/longitude (environ ~50 m)
    """
    results = []
    cars = df["CAR_ID"].unique()

    for car in cars:
        car_df = df[df["CAR_ID"] == car].sort_values("TIMESTAMP")
        starts = car_df[car_df["EVENT_TYPE"] == "ENGINE_START"]
        stops  = car_df[car_df["EVENT_TYPE"] == "ENGINE_STOP"]

        # Assure mÃªme nombre de start/stop (sinon anomalie)
        n = min(len(starts), len(stops))

        for i in range(n):
            s = starts.iloc[i]
            e = stops.iloc[i]

            lat_diff = abs(s["LATITUDE"]  - e["LATITUDE"])
            lon_diff = abs(s["LONGITUDE"] - e["LONGITUDE"])

            ok = lat_diff < tolerance and lon_diff < tolerance

            results.append({
                "CAR_ID": car,
                "TRIP_INDEX": i + 1,
                "START_TIME": s["TIMESTAMP"],
                "STOP_TIME": e["TIMESTAMP"],
                "START_LAT": s["LATITUDE"],
                "START_LON": s["LONGITUDE"],
                "STOP_LAT": e["LATITUDE"],
                "STOP_LON": e["LONGITUDE"],
                "LAT_DIFF": lat_diff,
                "LON_DIFF": lon_diff,
                "MATCH": ok
            })

    return pd.DataFrame(results)


# ðŸ” Appliquer le contrÃ´le
check_df = check_start_stop(df)

print("ðŸš¦ VÃ©rification Start=Stop pour toutes les voitures :")
display(check_df)

# Afficher uniquement anomalies
anomalies = check_df[check_df["MATCH"] == False]

if len(anomalies) == 0:
    print("âœ… Toutes les voitures reviennent bien Ã  leur position initiale.")
else:
    print("âŒ Anomalies dÃ©tectÃ©es :")
    display(anomalies)
```

```python
import pandas as pd

# 1) VÃ©rifier qu'on a folium
try:
    import folium
    from folium.plugins import AntPath
except ImportError as e:
    print("âŒ Le module 'folium' n'est pas installÃ©.")
    print("âž¡ Installe-le d'abord avec :  !pip install folium")
    raise e

# 2) VÃ©rifier que df existe
try:
    df
except NameError:
    raise NameError(
        "âŒ Le DataFrame 'df' n'existe pas. "
        "Charge d'abord tes CSV dans df (ex: df = load_all_telemetry())."
    )

# 3) VÃ©rifier les colonnes nÃ©cessaires
required_cols = {"CAR_ID", "LATITUDE", "LONGITUDE", "TIMESTAMP", "EVENT_TYPE"}
missing = required_cols - set(df.columns)
if missing:
    print("Colonnes prÃ©sentes :", list(df.columns))
    raise KeyError(f"âŒ Colonnes manquantes dans df : {missing}")

# 4) Nettoyage de base
telemetry = df.dropna(subset=["LATITUDE", "LONGITUDE"]).copy()
telemetry = telemetry.sort_values("TIMESTAMP")

if telemetry.empty:
    raise ValueError("âŒ Aucun point avec LATITUDE/LONGITUDE non nuls dans df.")

# 5) Choisir automatiquement une voiture qui a des points GPS
cars_counts = telemetry.groupby("CAR_ID").size().sort_values(ascending=False)
car_id = int(cars_counts.index[0])  # voiture avec le plus de points
print(f"ðŸš— CAR_ID sÃ©lectionnÃ© automatiquement : {car_id} ({cars_counts.iloc[0]} points)")

car_df = telemetry[telemetry["CAR_ID"] == car_id].copy()

# (Optionnel) filtrer sur 1 jour pour plus de lisibilitÃ©
# dÃ©commenter et adapte DATE_TARGET si tu veux :
# DATE_TARGET = pd.to_datetime("2025-12-10").date()
# car_df = car_df[car_df["TIMESTAMP"].dt.date == DATE_TARGET]

if car_df.empty:
    raise ValueError("âŒ car_df est vide aprÃ¨s filtrage. VÃ©rifie le filtre de date Ã©ventuellement.")

print(f"ðŸ“Š Nombre de points pour CAR_ID={car_id}: {len(car_df)}")

# 6) PrÃ©paration des points GPS
points = list(zip(car_df["LATITUDE"], car_df["LONGITUDE"]))

# Centre de la carte
center_lat = car_df["LATITUDE"].mean()
center_lon = car_df["LONGITUDE"].mean()

# 7) CrÃ©ation de la carte
m = folium.Map(location=[center_lat, center_lon], zoom_start=12)

# Trajet continu
folium.PolyLine(
    locations=points,
    weight=4,
    opacity=0.8,
).add_to(m)

# Variante animÃ©e (ligne "vivante")
AntPath(
    locations=points,
    dash_array=[10, 20],
    delay=800,
).add_to(m)

# 8) Marqueurs START/STOP

start_point = car_df.iloc[0]
end_point   = car_df.iloc[-1]

folium.Marker(
    location=[start_point["LATITUDE"], start_point["LONGITUDE"]],
    popup=f"START<br>{start_point['TIMESTAMP']}",
    icon=folium.Icon(color="green", icon="play")
).add_to(m)

folium.Marker(
    location=[end_point["LATITUDE"], end_point["LONGITUDE"]],
    popup=f"STOP<br>{end_point['TIMESTAMP']}",
    icon=folium.Icon(color="red", icon="stop")
).add_to(m)

print("âœ… Carte gÃ©nÃ©rÃ©e. Si tu es dans un notebook, la carte doit s'afficher juste en dessous ðŸ‘‡")

m
```
